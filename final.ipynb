```python
import cv2
from PIL import Image, ImageEnhance
from matplotlib import pyplot as plt
```

```python
#
img = Image.open("img_cat.jpg")

# 
brightness = ImageEnhance.Brightness(img)

output_a = brightness.enhance(5)         # 
output_b = brightness.enhance(0.5)      # 

plt.figure(figsize=(20,15))
plt.subplot(2,4,1)
plt.imshow(img)
plt.title('Original')
plt.subplot(2,4,2)
plt.imshow(output_a)
plt.title('brightness:5')
plt.subplot(2,4,3)
plt.imshow(output_b)
plt.title('brightness:0.5')
```

```python
#
img = Image.open("img_cat.jpg")

# 
contrast = ImageEnhance.Contrast(img)

output_a = contrast.enhance(5)           # 
output_b = contrast.enhance(0.5)        # 

plt.figure(figsize=(20,15))
plt.subplot(2,4,1)
plt.imshow(img)
plt.title('Original')
plt.subplot(2,4,2)
plt.imshow(output_a)
plt.title('contrast:5')
plt.subplot(2,4,3)
plt.imshow(output_b)
plt.title('contrast:0.5')
```
